```python
def decode(individual):
    sorted_individual = sorted(individual,reverse=True)
    sorted_index = list()
    for i in range(len(individual)):
        sorted_index.append(individual.index(sorted_individual[i]))
    return sorted_index
```

## Phase2: Selection for breeding offsprings
- Test fitness and select individuals to create a parent population
- Reproduction by crossover
- Reproduction by mutation

```python
def fitness1(individual):
    '''minus distance'''
    fitness_value = 0
    r = decode(individual)
    for i in range(len(individual)):
        #print('Distance from {0} to {1} = {2}'.format(Province[r[i%len(individual)]],Province[r[(i+1)%len(individual)]],Distance[r[i%len(individual)],r[(i+1)%len(individual)]]))
        fitness_value += Distance[r[i%len(individual)],r[(i+1)%len(individual)]]
    #print('Total {}'.format(fitness_value))
    return fitness_value* -1

def fitness2(individual):
    '''Different of distance from max'''
    fitness_value = 0
    total = 0
    max_distance = sum([max(Distance[i]) for i in range(len(Distance))])
    r = decode(individual)
    P = [Province[r[i%len(individual)]] for i in range(len(individual))]
    D = [Distance[r[i%len(individual)],r[(i+1)%len(individual)]] for i in range(len(individual))]
    total = sum([Distance[r[i%len(individual)],r[(i+1)%len(individual)]] for i in range(len(individual))])
    
    fitness_value = sum([max_distance - Distance[r[i%len(individual)],r[(i+1)%len(individual)]] for i in range(len(individual))])
    #print('Max_distance'.format(max_distance))
    #print('Province {}'.format(P))
    #print('Distance {}'.format(D))
    #print('Total Distance {}'.format(total))
    #print('Fitness_value {}'.format(fitness_value))
    return fitness_value

def fitness3(individual):
    '''Power of different of distance from max'''
    fitness_value = 0
    total = 0
    max_distance = sum([max(Distance[i]) for i in range(len(Distance))])
    r = decode(individual)
    P = [Province[r[i%len(individual)]] for i in range(len(individual))]
    D = [Distance[r[i%len(individual)],r[(i+1)%len(individual)]] for i in range(len(individual))]
    total = sum([Distance[r[i%len(individual)],r[(i+1)%len(individual)]] for i in range(len(individual))])
    fitness_value = sum([(max_distance - Distance[r[i%len(individual)],r[(i+1)%len(individual)]])**2 for i in range(len(individual))])
    #print('Max_distance'.format(max_distance))
    #print('Province {}'.format(P))
    #print('Distance {}'.format(D))
    #print('Total Distance {}'.format(total))
    #print('Fitness_value {}'.format(fitness_value))
    return fitness_value
```

```python
import numpy.random as npr
def roulette_wheel_selection(population):
    max = sum([fitness1(pop) for pop in population])
    selection_p = [fitness1(pop)/max for pop in population]
    select = npr.choice(len(population),p = selection_p)
    print('Select population[{0}]: {1} with probabily:{2:.2f} '.format(select,population[select],selection_p[select]))
    #print('Select population[{0}]: {1} with probabily:{2:.2f} '.format(select,decode(population[select]),selection_p[select]))
    return population[select]
```

```python
def select_parent_population(population,k):
    selected_parents = list()
    for i in range(k):
        selected_parent = roulette_wheel_selection(population)
        selected_parents.append(selected_parent)
    return selected_parents

def make_unique(chromosome):
    for i in range(len(chromosome)-1):
        for j in range(i+1,len(chromosome)):
            if (chromosome[i] == chromosome[j]):
                chromosome[i] += chromosome.count(chromosome[i]) -1
                print('Change chromosome at position[{0}] to {1}'.format(i,chromosome[i]))
    return chromosome

def crossover(parent1,parent2):
    parent1 = list(parent1)
    parent2 = list(parent2)
    cp = random.randint(1,len(parent1))# no clone allowed
    # crossover point
    # cp = 0 [|0 1 2 3 4 5] - clone
    # cp = 1 [0|1 2 3 4 5]
    # cp = 2 [0 1|2 3 4 5]
    # cp = 3 [0 1 2|3 4 5]
    # cp = 4 [0 1 2 3|4 5]
    # cp = 5 [0 1 2 3 4|5]
    # cp = 6 [0 1 2 3 4 5|] - clone
    print('Crossover point: {}'.format(cp))
    print('Parent1: {0} x Parent2: {1}'.format(parent1,parent2))
    #print('Parent1: {0} x Parent2: {1}'.format(decode(parent1),decode(parent2)))
    offspring1,offspring2 = parent1[:cp],parent2[:cp]
    offspring1.extend(parent2[cp:])
    offspring2.extend(parent1[cp:])
    offspring1 = make_unique(offspring1)
    offspring2 = make_unique(offspring2)
    print('Offspring1: {0} , Offspring2: {1}'.format(offspring1,offspring2))
    #print('Offspring1: {0} , Offspring2: {1}'.format(decode(offspring1),decode(offspring2)))
    return offspring1,offspring2

def mutation(individual):
    '''Randomly swap position'''
    mutated_ind = list()
    mp1 = random.randint(0,len(individual))
    mp2 = random.randint(0,len(individual))
    mutated_ind[:] = individual[:]
    mutated_ind[mp1],mutated_ind[mp2] = mutated_ind[mp2],mutated_ind[mp1]
    print('Swap position {0} and {1} = Mutated: {2}'.format(mp1,mp2,mutated_ind))
    #print('Swap position {0} and {1} = Mutated: {2}'.format(mp1,mp2,decode(mutated_ind)))
    return mutated_ind

def breeding_offspring(population,num_cross,num_mutate):
    #Crossover operation
    print('------ Crossover -------')
    selected_parents = select_parent_population(population,num_cross)
    for i in range(int(num_cross/2)):
        offspring1, offspring2 = crossover(selected_parents[i*2],selected_parents[i*2+1])
        population.extend([offspring1,offspring2])
    #Mutantation operation
    print('------ Mutation -------')
    mutated_inds = list()
    for i in range(num_mutate):
        rand_ind = random.randint(0,len(population))
        print('Mutating Population[{0}]: {1}'.format(rand_ind,(population[rand_ind])))
        mutated_ind = mutation(population[rand_ind])
        mutated_inds.append(mutated_ind)
    population.extend(mutated_inds)
    print('-----------------------')
```

## Phase3: Evaluation of a population
- Values of the fitness function are computed for individuals of the new offspring population
- Replace the old generation with the higher value fitness offsprings
- Terminate when the optimum solution are committed, or numbers of iteration  are reached

```python
def update_population(population,num_pop):
    for i in range(len(population)):
        print('population[{0}]: {1} fitness: {2}'.format(i,population[i],fitness1(population[i])))
        #print('population[{0}]: {1} fitness: {2}'.format(i,decode(population[i]),fitness1(population[i])))
        
    num_kill_pop = len(population) - num_pop
    evaluation = [fitness1(population[i]) for i in range(len(population))]
    for i in range(num_kill_pop):
        kill_pop = evaluation.index(min(evaluation))
        print('kill {0} with fitness: {1}'.format(population[kill_pop],evaluation[kill_pop]))
        #print('kill {0} with fitness: {1}'.format(decode(population[kill_pop]),evaluation[kill_pop]))
        evaluation.remove(min(evaluation))
        population.remove(population[kill_pop])
    max_fitness_value = max(evaluation)
    max_fit_pop = population[evaluation.index(max(evaluation))]
    return max_fit_pop,max_fitness_value
```

```python
max_fit_pop, max_fitness_value = update_population(population,10)
```

# Genetic Algorithm Run

```python
num_iteration = 10
num_pop = 20
num_cross = 6
num_mutate = 2
# Generate Population Generation[0]
population = generatePop(num_pop)
# Loop for breeding and update population
for iter in range(num_iteration):
    print('- Generation[{0}] -'.format(iter))
    for i in range(len(population)):
        print('population[{0}]: {1} fitness: {2}'.format(i,population[i],fitness1(population[i])))
        #print('population[{0}]: {1} fitness: {2}'.format(i,decode(population[i]),fitness1(population[i])))    
    breeding_offspring(population,num_cross,num_mutate)
    max_fit_pop,max_fitness_value = update_population(population,num_pop)
    
    # display result at Generation[i]
    s = decode(max_fit_pop)
    provinceList = [Province[s[i]] for i in range(len(population[0]))]
    total_distance = fitness1(max_fit_pop)*-1
    print('- End Generation[{0}]: Result:{1} with fitness: {2} -'.format(iter,max_fit_pop,max_fitness_value))
    #print('- End Generation[{0}]: Result:{1} with fitness: {2} -'.format(iter,decode(max_fit_pop),max_fitness_value))
    print('- Best path:{0} with distance {1} km -'.format(provinceList,total_distance))
    print('------------------------------------------------------------------')
```
