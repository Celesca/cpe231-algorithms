### Experimental 1 - Recursive and non-recursive algorithm for computing n!

```python
import time
import sys

# üëáÔ∏è set recursion limit to 2000
sys.setrecursionlimit(50000)
```

```python
# Recursive Factorial

def recursive_factorial(n):
    if n == 0:
        return 1
    else:
        return n * recursive_factorial(n - 1)
    
batch_epoch = [int(i) for i in range(1,20001,1000)]

re_factorial_time_list = []

for number in batch_epoch:

    start = time.time()

    result = recursive_factorial(number)

    end = time.time()

    re_factorial_time_list.append(end-start)

    print("Time of recursive_factorial" , number , ":" , end-start , "s")

print(re_factorial_time_list)
    
```

```python
# Non-Recursive Factorial

def nonRecursive_factorial(n):
    if n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result
    
batch_epoch = [int(i) for i in range(1,20001,1000)]
nonRe_factorial_time_list = []
    
for number in batch_epoch:

    start = time.time()

    result = nonRecursive_factorial(number)

    end = time.time()

    nonRe_factorial_time_list.append(end-start)

    print("Time of Non-Recursive_factorial" , number , ":" ,end-start , "s")

print(nonRe_factorial_time_list)
```

```python
import matplotlib.pyplot as plt

plt.xlabel("Input (n)")
plt.ylabel("time (s)")

plt.plot(batch_epoch,re_factorial_time_list, label="Recursive Factorial")
plt.plot(batch_epoch, nonRe_factorial_time_list, label="Non-Recursive Factorial")
plt.legend()
plt.show()
```

```python
def recursive_fibonacci(n):
    if n < 1:
        return n
    else:
        return recursive_fibonacci(n-1) + recursive_fibonacci(n-2)
    
    
batch_epoch = [int(i) for i in range(1,41)]

re_fibonacci_time_list = []

for number in batch_epoch:

    start = time.time()

    result = recursive_fibonacci(number)

    end = time.time()

    re_fibonacci_time_list.append(end-start)

    print("Time of recursive_factorial" , number , ":" , end-start , "s")

print(re_fibonacci_time_list)
```

```python
def nonRecursive_fibonacci(n):

    # Declare the local variable
    curr_num = n 
    x = 1
    y = 0

    while (curr_num >= 0) :
        temp = x
        x = x + y
        y = temp
        curr_num -= 1
    
    return y
    
batch_epoch = [int(i) for i in range(1,41)]

nonRe_fibonacci_time_list = []

for number in batch_epoch:

    start = time.time()

    result = nonRecursive_fibonacci(number)

    end = time.time()

    nonRe_fibonacci_time_list.append(end-start)

    print("Time of recursive_factorial :" , end-start , "s")

print(nonRe_fibonacci_time_list)
```

```python
import matplotlib.pyplot as plt

plt.xlabel("Input (n)")
plt.ylabel("time (s)")

plt.plot(batch_epoch, re_fibonacci_time_list, label="Recursive Fibonacci")
plt.plot(batch_epoch, nonRe_fibonacci_time_list, label="Non-Recursive Fibonacci")
plt.legend()
plt.show()
```
